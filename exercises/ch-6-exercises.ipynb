```python
#----------R6-3---------------------
class Empty(Exception):
    pass

class Stack():
    def __init__(self):
        self._data = []
    
    def __len__(self):
        return len(self._data)
    
    @property
    def data(self):
        return self._data
    
    def is_empty(self):
        return len(self._data) == 0
    
    def push(self, value):
        self._data.append(value)
    
    def top(self):
        return self._data[-1]
        
    def pop(self):
        if self.is_empty():
            raise Empty('List is empty')
        return self._data.pop()
    
    
    def full_pop(self):
        ans = []
        while not self.is_empty():
            ans.append(self.pop())
        return ans
    
    def transfer_ST(self, T):
        for _ in range(len(self._data)):
            T.push(self.pop())
        
        

def transfer(S, T):
    while not S.is_empty():
        T.push(S.pop())

S,T = Stack(), Stack()

try: S.pop()
except Exception as e: print (e)
    
    
for i in range(5,20):
    S.push(i)
T.push(0)
print('Top of S is: ', S.top())
print('Top of T is before: ', T.top())
transfer(S, T)
print('Top of T is after: ', T.top())

print('====> Using transfer_ST')
xS = Stack()
for i in range(1,5):
    xS.push(i)
print('S ', xS.data)
yT = Stack()
xS.transfer_ST(yT)
print('After transfer')
print('S ', xS.data)
print('T', yT.data)
```
