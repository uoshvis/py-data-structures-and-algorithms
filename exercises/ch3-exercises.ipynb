```python
#-----------R3-1-----------------
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = [10**i for i in range(10)]
funcs = [lambda x: 8*x,
        lambda x: 4*x*math.log(x),
        lambda x: 2*x**2,
        lambda x: x**3,
        #lambda x: 2**x    #Note, this is too time consuming to compute
        ]
ys = []
for func in funcs:
    ys.append(list(map(func, x)))

for y in ys:
    plt.plot(x, y)
plt.yscale('log')
plt.xscale('log')
```

```python
#-------------R3-2-------------------
x = [x/100 for x in range(1, 500)]
y1 = list(map(lambda x: 4*math.log(x), x))
y2 = x
plt.plot(x, y1)
plt.plot(x, y2)
```

```python
# R-3.23
def example1(S):
    """Return the sum of the elements in sequence S."""
    n = len(S)
    total = 0
    count = 0
    num_calc = []
    for j in range(n):
        total += S[j]
        count += 1
        num_calc.append(count)
    return total, num_calc
```

```python
s = np.random.randint(0, 100, size=(100,))
```

```python
total, num_calc = example1(s)
```

```python
plt.plot(num_calc)
```
